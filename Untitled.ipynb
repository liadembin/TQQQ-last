```python
data_arr = IndicatorData.to_numpy()
```

```python
X = []
y = []
for i in range(0,len(data_arr)-window_size):
    X.append(data_arr[i:i+window_size])
    y.append(data_arr[i+window_size,0])
X = np.array(X)
y = np.array(y)
print("X shape: ",X.shape)
print("y shape: ",y.shape)
```

```python
from sklearn.model_selection import train_test_split
X = X.astype("float64")
y = y.astype("float64")
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1-trainTestSplit, random_state=8)
```

```python
model = Sequential()
```

```python
model.add(Bi(LSTM(512,return_sequences=True,input_shape=(X_train[0].shape))))
model.add(Dropout(.2))
```

```python
model.add(Bi(LSTM(256,return_sequences=True)))
model.add(Dropout(.2))
```

```python
model.add(Bi(LSTM(64,return_sequences=False)))
model.add(Dropout(.2))
```

```python
model.add(Dense(64,activation='relu'))
model.add(Dense(1,activation='linear'))
```

```python
from tensorflow.keras.metrics import mean_squared_logarithmic_error ,RootMeanSquaredError,MeanAbsolutePercentageError  
model.compile(optimizer='rmsprop',loss='mse',metrics=['mae',RootMeanSquaredError(),MeanAbsolutePercentageError()])
```

```python
import datetime
log_dir = "logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=log_dir, histogram_freq=1)
model.fit(X_train,y_train,shuffle=True,validation_data=(X_test,y_test),epochs=100,use_multiprocessing=True,callbacks=[tensorboard_callback])
```
